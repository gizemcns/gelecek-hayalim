```python
#5
#yukarıda bulduğumuzu görselleştirelim.

race = df["race/ethnicity"].value_counts()

plt.figure(figsize=(6,4))
plt.bar(race.index, race.values)
plt.title("Grup Sayı Dağılımı")
plt.xlabel("Grup İsmi")
plt.ylabel("Kişi Sayısı")
plt.show()
```
